```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

main_dir = "/home/nadav2/dev/data/CLIP2Shape/outs/descriptors_ablations/smplx_male_03"

to_compare_dirs = ["smplx_male_2_descriptors", "smpl_male_new_algorithm", "smplx_male_5_descriptors", "smplx_male_10_descriptors", "smplx_male_15_descriptors", "smplx_male_20_descriptors", "smplx_male_25_descriptors"]
production_dir = "smpl_male_new_algorithm"
results_dict = {"iou": [], "num_of_descriptors": [], "overlap": []}
for to_compare_dir in to_compare_dirs:
    with open(Path(main_dir) / to_compare_dir /  "descriptors.json", "r") as f:
        data = json.load(f)
    results_dict["iou"].append(data["iou"]  * 100)
    results_dict["num_of_descriptors"].append(data["labels"].__len__())
    results_dict["overlap"].append(data["overlap"] * 100)
    if to_compare_dir == production_dir:
        production_iou = data["iou"] * 100
        production_num_of_descriptors = data["labels"].__len__()
        production_overlap = data["overlap"] * 100


# plot the results, x axis is the number of descriptors, y axis is the iou
plt.figure(figsize=(10, 5))
plt.plot(results_dict["num_of_descriptors"], results_dict["iou"], label="vertices coverage", color="blue")
plt.plot(results_dict["num_of_descriptors"], results_dict["overlap"], label="overlap", color="green")
plt.xticks(results_dict["num_of_descriptors"].remove(production_num_of_descriptors))
plt.yticks(np.arange(0, 101, 20), labels=[f"{i}%" for i in np.arange(0, 101, 20)])
plt.axvline(x=production_num_of_descriptors, color="red", linestyle="--", alpha=0.3, label="without limit")
plt.annotate(f"{production_num_of_descriptors} descriptors\n{production_iou:.2f}% vertices coverage\n{production_overlap:.2f}% overlap", (production_num_of_descriptors+0.5, production_iou+7.0))
plt.xlabel("# Descriptors")
plt.legend()
plt.title("Performance by # Descriptors - male")
plt.tight_layout()

# save the plot
plt.savefig(Path(main_dir) / "descriptors_vs_coverage.png")
```
